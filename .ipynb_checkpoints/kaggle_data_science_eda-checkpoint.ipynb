Pull in the survey response data, the free form response data, and the schema:

```python
# Multiple choice responses
df_mc = pd.read_csv('multipleChoiceResponses.csv')
# Free form responses
df_ff = pd.read_csv('freeFormResponses.csv')
# Survey schema
df_sc = pd.read_csv('SurveySchema.csv')
```

It looks like this dataset has a lot of null value columns, let's get an exact count:

```python
df_mc_low_nulls = df_mc.columns[np.sum(df_mc.isnull())/df_mc.shape[0] < 0.95]
df_mc_low_nulls
```

It looks as though only 236 out of the 393 columns have less than 95% of their values equal to null. We'll remove all columns with more than 95% null values:

```python
df_mc.shape
```

```python
#As an initial data cleaning step, let's remove the 'Time to finish' responses:
df_mc = df_mc.drop(df_mc.columns[1],axis=1)
df_mc
```

Question: what can we find out about companies who use machine learning regularly vs those who don't?

```python
ml_methods = df_mc['Q10'].value_counts().drop(labels='Does your current employer incorporate machine learning methods into their business?')
ml_methods
```

```python

ml_methods.plot(kind='pie',autopct='%1.2f',title=None)
```
